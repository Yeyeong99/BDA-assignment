```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
observed = np.array([[60,40,30,20], [45,20,30,25]])
```




### 선호도 데이터

|성별|아메리카노|카페라떼|카푸치노|아이스티|
|:-:|:-:|:-:|:-:|:-:|
여성|60|40|30|20
남성|45|20|30|25


----


```python
observed
```

```python
chi2, p, dof, expected = chi2_contingency(observed)
```

```python
print(f"카이제곱 검정통계량:,{chi2:.4f}")
print(f"p값:,{p:.4f}")
print(f"자유도:,{dof}")
print(f"기대빈도:,{expected}")
```

- 카이제곱 검정 통계량
    - 관측값과 기대값의 사이의 차이를 측정한 값
    - 예) 공정한 주사위는 1부터 6까지 나올 확률이 1/6 기대(기대값) 실제로 실행했을 때의 값(관측값)사이 차이는 얼마인지 측정한 것
- p값
    - **귀무가설이 참이라는 가정 하에** 관측된 데이터가 나올 확률
- 자유도
    - 독립적인 차원 수 n-1
- 기대빈도
    - 귀무가설이 참일 때 예상되는 빈도
    - 아메리카노
        - 여성 총합(150)X아메리카노총합(105)/ 전체총합(270)
