```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# 랜덤하게 수를 생성. 시드를 고정해 결과를 동일하게 재현하도록 함.
np.random.seed(42)

n_trials= 100
p_fair = 0.5 #게임 승리 확률
p_suspicious = 0.7 #의심스럽게 승부조작을 하게 되어 이길 확률

# 예제데이터 생성
# binom.rvs: 이항 분포 상 난수를 생성
# 100번의 시행에서 성공 확률이 p_fair(50%)인 1000개의 난수를 생성
fair_data=binom.rvs(n = n_trials, p= p_fair, size=1000)

# fair_data와 동일, 성공 확률만 p_suspicious = 0.7인 데이터
suspicious_data=binom.rvs(n = n_trials, p= p_suspicious, size=1000)

# 가능한 승리의 횟수
x = np.arange(0, n_trials+1)

# binom.pmf: x값에 대하여 특정 게임에서 승리할 확률 계산
fair_prob =binom.pmf(x, n_trials, p_fair)
suspicious_prob =binom.pmf(x, n_trials, p_suspicious)
```

```python
fair_prob
```

```python
suspicious_prob
```

```python
plt.figure(figsize=(12,6))
plt.bar(x, fair_prob, width= 0.4, label= 'fair', alpha=0.7)
plt.bar(x, suspicious_prob, width= 0.4, label= 'suspicious', alpha=0.7)
plt.title('fair vs suspicious')
plt.legend()
plt.show()
```

```python
# 누적 분포함수 (CDF) Comulative Distribution Function

fair_cdf =binom.cdf(x, n_trials, p_fair)
suspicious_cdf =binom.cdf(x, n_trials, p_suspicious)
```

```python
plt.figure(figsize=(12,6))
plt.bar(x, fair_cdf, label= 'fair' )
plt.bar(x, suspicious_cdf, label= 'suspicious')
plt.title('fair vs suspicious')
plt.legend()
plt.show()
```
