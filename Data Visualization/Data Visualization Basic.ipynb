```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

# 데이터 생성
np.random.seed(42)
# 정규분포
data_normal = np.random.normal(0, 2, 1000)
# 양의 왜도
data_positive_skew = np.random.exponential(5, 1000)
# 음의 왜도
data_negative_skew = -np.random.exponential(5, 1000)
# 높은 첨도, T-분포
data_high_kurtosis = np.random.standard_t(10, 1000)
# 낮은 첨도, 균등 분포
data_low_kurtosis = np.random.uniform(-2, 4, 1000)

# 데이터와 통계 계산
datasets = {
    "normal": data_normal,
    "positive skew": data_positive_skew,
    "negative skew": data_negative_skew,
    "high kurtosis": data_high_kurtosis,
    "low kurtosis": data_low_kurtosis,
}
```

```python
#시각화
fig,axes = plt.subplots(3, 2, figsize = (10,10))
axes = axes.flatten()

for i, (name, data) in enumerate(datasets.items()):
    skewness = skew(data)
    kurt = kurtosis(data, fisher = False) #기준값이 3
    x = np.linspace(np.min(data), np.max(data), 1000)
    y = norm.pdf(x, np.mean(data), np.std(data))

    axes[i].hist(data, bins = 20, alpha = 0.7, density = True)
    axes[i].plot(x, y, label = 'Normal Curve', color = 'red')
    axes[i].legend()

#레이아웃 조정
plt.tight_layout()
plt.show()
```
