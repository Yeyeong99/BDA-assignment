```python
import seaborn as sns
tt= sns.load_dataset('titanic')
```

```python
# 성별 생존율
ss_gr=tt.groupby(['sex'])
```

```python
# 남성 생존율
survived_male = ss_gr.get_group('male')['survived'].mean() *100
```

```python
#여성 생존율
survived_female = ss_gr.get_group('female')['survived'].mean() *100
```

```python
#생존율 차
survived_female - survived_male
```

```python
fare_male = ss_gr.get_group('male')['fare'].mean()
```

```python
fare_female = ss_gr.get_group('female')['fare'].mean()
```

```python
fare_male - fare_female
```

```python
ss_gr.get_group('female')[['survived', 'fare']].describe()
```

```python
ss_gr.get_group('male')[['survived', 'fare']].describe()
```

총 수는 남성이 더 많지만 생존율과 비용 요금 평균 모두 여성이 높음을 확인할 수 있었다.

```python
#클래스 별 군집
gr_pc=tt.groupby('pclass')
```

```python
#클래스 별 생존율
survived_1 = gr_pc.get_group(1)['survived'].mean() *100
survived_2 = gr_pc.get_group(2)['survived'].mean() *100
survived_3 = gr_pc.get_group(3)['survived'].mean() *100
```

```python
survived_1, survived_2, survived_3
```

```python
display('pclass1 survived fare 기초 통계',gr_pc.get_group(1)[['survived', 'fare']].describe())
display('pclass2 survived fare 기초 통계',gr_pc.get_group(2)[['survived',  'fare']].describe())
display('pclass3 survived fare 기초 통계',gr_pc.get_group(3)[['survived', 'fare']].describe())
```

클래스가 높을 수록 수는 적지만 생존율은 높은 것으로 나타난다.
