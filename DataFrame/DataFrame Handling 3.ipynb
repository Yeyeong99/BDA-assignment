혼자가 아닌 경우가 전반적으로 생존율이 높은 것으로 나타난다.

```python
new_filter=tt[(tt['age'] <20)][['fare','survived']]
```

```python
new_filter2=tt[(tt['age'] >=20)][['fare','survived']]
```

```python
new_filter[['survived', 'fare']].describe()
```

```python
new_filter2[['survived', 'fare']].describe()
```

20세 미만인 경우 20세 이상인 경우보다 생존율이 10%가량 높은 것으로 나타났다.
