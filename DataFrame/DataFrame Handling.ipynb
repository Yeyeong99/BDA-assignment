```python
tt_sp=tt[['pclass','age','fare']]

pc_gr=tt_sp.groupby('pclass')
```

```python
pc_gr.mean()
```

```python
pc_gr.count()
```

```python
pc_gr.std()
```

```python
## 다양한 통계치 = agg / mode는 없다
tt.groupby('pclass').agg(fare_sum = ('fare','sum'),
                        fare_mode= ('fare','mode'),
                        n_cnt = ('fare','count'))
```

```python
## agg 함수는 내가 만든 함수도 적용시킬 수 있다!

def min_max(x):
    return x.max() - x.min()
```

```python
#내가 만든 튜닝한 함수를 agg 넣어서 바로 원하는 값을 출력할 수 있다.
pc_gr.agg(min_max)
```

```python
gr_pc=tt.groupby('class')
```

```python

## 기초통계치 확인
display('class First 기초통계',gr_pc.get_group('First').describe())
display('class Second 기초통계',gr_pc.get_group('Second').describe())
display('class Third 기초통계',gr_pc.get_group('Third').describe())
```

```python
## 생존율 계산

display('class First 생존율',gr_pc.get_group('First')['survived'].mean() * 100)
display('pclass Second 생존율',gr_pc.get_group('Second')['survived'].mean() * 100)
display('pclass Third 생존율',gr_pc.get_group('Third')['survived'].mean() * 100)
```
