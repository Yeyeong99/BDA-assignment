## 주요 키워드 정리

### 1. **binom.pmf(k, n, p)**
- **특정 횟수 k에서의 확률 질량 함수(PMF)를 계산.**
- **파라미터:**
  - k: 성공 횟수 (정수 | 배열)
  - n: 실험 횟수 (정수)
  - p: 성공 확률 (0 <= p <= 1)
- **예:**
  ```python
  from scipy.stats import binom
  prob = binom.pmf(k=5, n=10, p=0.5)  # 10번 중 5번 성공할 확률
  ```
---
### 2. **binom.cdf(k, n, p)**
- **특정 횟수 `k` 이하의 누적 확률 계산.**  
- **파라미터:**
  - k: 성공 횟수 (정수 또는 배열).
  - n: 실험 횟수 (정수).
  - p: 성공 확률 (0 <= p <= 1).
- **예:**
  ```python
  cdf = binom.cdf(k=5, n=10, p=0.5)  # 10번 중 5번 이하로 성공할 확률
  ```

---

### 3. **binom.sf(k, n, p)**
- **특정 횟수 k 이상의 누적 확률 계산 (Survival Function).**
- **파라미터:**
  - k: 성공 횟수 (정수 또는 배열).
  - n: 실험 횟수 (정수).
  - p: 성공 확률 (0 <= p <= 1).
- **예:**
  ```python
  sf = binom.sf(k=5, n=10, p=0.5)  # 10번 중 5번 이상 성공할 확률
  ```

---

### 4. **binom.rvs(n, p, size)**
- **이항분포에서 랜덤 변수 생성.**
- **파라미터:**
  - n: 실험 횟수 (정수).
  - p: 성공 확률 (0 <= p <= 1).
  - size: 생성할 샘플 크기.
- **예:**
  ```python
  samples = binom.rvs(n=10, p=0.5, size=1000)  # 10번 실험을 1000번 반복한 샘플 생성
  ```

---

### 5. **`binom.stats(n, p, moments)`**
- **이항분포의 통계적 속성(평균, 분산, 왜도, 첨도) 계산.**
- **파라미터:**
  - `n`: 실험 횟수.
  - `p`: 성공 확률.
  - `moments`: 문자열로 통계량 지정 (`'mvsk'` = 평균, 분산, 왜도, 첨도).
- **예:**
  ```python
  mean, var, skew, kurt = binom.stats(n=10, p=0.5, moments='mvsk')
  ```

---

### 6. **binom.mean(n, p)**

### 7. **binom.var(n, p)**

### 9. **binom.interval(alpha, n, p)**
- **주어진 신뢰 구간에서 성공 횟수의 범위를 계산.**
- **파라미터:**
  - alpha: 신뢰 수준 (0 < alpha < 1).
  - n: 실험 횟수.
  - p: 성공 확률.
- **예:**
  ```python
  ci = binom.interval(0.95, n=10, p=0.5)  # 95% 신뢰 구간
  ```

#이항분포 과제 실습

````python
from scipy.stats import binom
from matplotlib import pyplot as plt
import numpy as np

wins = np.arange(0, 21) # 가능한 승리의 횟수
prob = np.array([])
for win in wins:
  prob = np.append(prob, binom.pmf(k=win, n=20, p=0.5))
print(prob)
# prob 확률질량함수 plot
plt.figure(figsize=(5,5))
plt.bar(wins, prob, width= 0.4, label= 'fair', alpha=0.7)
plt.show()
````

````python
wins_2 = np.arange(0, 21) # 가능한 승리의 횟수
prob_list_2 = np.arange(0.1, 1.0, 0.2)

plt.figure(figsize = (5,5))
for p in prob_list_2:
#확률밀도함수 plot
  cdf = binom.cdf(k=wins_2, n=20, p=p)
  plt.bar(wins_2 - 0.5 * p, cdf, label = str(round(p,1)), width = 0.5, alpha = 0.4)
plt.xlabel('number of wins')
plt.ylabel('probability')

plt.legend()
plt.show()
````

````python
wins_3 = np.arange(0, 21) # 가능한 승리의 횟수
prob_list_3 = np.arange(0.1, 1.0, 0.2)

plt.figure(figsize = (5,5))
for p in prob_list_3:
#확률밀도함수 plot
  sf = binom.sf(k=wins_3, n=20, p=p)
  plt.bar(wins_3 - 0.5 * p, sf, label = str(round(p,1)), width = 0.5, alpha = 0.4)
plt.xlabel('number of wins')
plt.ylabel('probability')

plt.legend()
plt.show()
````

````python
ci = binom.interval(0.95, n=20, p=0.5)  # 95% 신뢰 구간
ci
````

##왜도 (Skewness)
분포의 비대칭 정도, 왼쪽 혹은 오른쪽으로 치우침
##첨도 (Kurtosis)
중앙부분(평균)의 뾰족함의 정도를 나타내는 수치

```
정규분포가 어떠한 왜도와 첨도인지를 아는 경우, 해당 정규분포와 다른 수치는 기본적으로 정규분포가 아님을 알 수 있음
정규분포는 첨도가 0이 왜도 2~3
```
