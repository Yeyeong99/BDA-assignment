```python
import seaborn as sns
import pandas as pd
```

```python
# 타이타닉 데이터 로드
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# 등급 -> 성별 별로 그룹화 하여 통계 분석
grouped_by_sex = titanic.groupby('sex')
```

```python
female_stats = grouped_by_sex.get_group('female').describe()
male_stats = grouped_by_sex.get_group('male').describe()
```

```python
print(f'여성 탑승객 통계치:\n {female_stats}')
```

```python
print(f'남성 탑승객 통계치:\n {male_stats}')
```

```python
#생존률 계산 (전체 탑승객 수 대비 생존자 수)
female_survival_rate = grouped_by_sex.get_group('female')['survived'].mean() * 100
male_survival_rate = grouped_by_sex.get_group('male')['survived'].mean() * 100
```

```python
print(f'여성 생존률: \n {female_survival_rate}')
```

```python
print(f'남성 생존률: \n {male_survival_rate}')
```

```python
from scipy import stats
```

```python
# 타이타닉의 성별에 따른 생존자와 사상자 수를 계산
# 카이제곱 통계량 p-value, 자유도, 기대값 테이블 등을 이용해
# 통계적으로 성별에 따른 생존률 차이가 유의미한지 분석

crss_table = pd.crosstab(titanic['sex'], titanic['survived'])
```

```python
chi2, p, dof, expected =stats.chi2_contingency(crss_table)
```

```python
## 결과출력

print('카이제곱 통계량',chi2)
print()
print('p-value',p)
print()
print('자유도',dof)
print()
print('기댓값 테이블',expected)
print()

## p-value 결과 도출
if p < 0.05:
    print('성별에 따른 생존률 차이는 통계적으로 유의미하다.')
else:
    print('성별에 따른 생존률 차이는 통계적으로 유의미 하지 않다.')
```
