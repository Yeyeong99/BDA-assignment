```python
# 데이터 불러오기
train = pd.read_csv("/kaggle/input/bigdatacertificationkr/diabetes_train.csv")
test = pd.read_csv("/kaggle/input/bigdatacertificationkr/diabetes_test.csv")
train.shape, test.shape
```

```python
predict = rf.fit(X_train, y)
```

```python
result = rf.predict(test)
```

```python
result_df = pd.DataFrame({'pred': result})
```

## 데이터 전처리

```python
# 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
train[cols] = scaler.fit_transform(train[cols])
test[cols] = scaler.transform(test[cols])
```

```python
# id 제외
train = train.drop('id',axis=1)
test = test.drop('id',axis=1)
```

```python
# 타겟
target = train.pop("Outcome")
```

## 학습 및 예측

```python
# 학습 및 예측
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=2022)
model.fit(train, target)
predictions = model.predict_proba(test)
```

```python
predictions
```

```python
# 당뇨 확률값 선택
pred = predictions[:,1]
pred
```

## 파일 생성

```python
output = pd.DataFrame({'pred': pred})
output.head()
```

```python
# 수험번호.csv로 출력
output.to_csv('result.csv', index=False)
```
